```python
import schemdraw

import sdcomms as comms

schemdraw.use("matplotlib")


def drawElements(elm_list, n=5, dx=1, dy=2, ofst=0.8, fname=None, **kwargs):
    x, y = 0, 0
    d = schemdraw.Drawing(fontsize=16)
    for e in elm_list:
        element = getattr(comms, e)
        d += element().at(((d.unit + 0.5) * x + 0.5, y))
        x = x + dx
        if x >= n:
            x = 0
            y = y - dy
    return d


elems = [
    "Rectangle",
    "CouplerDot",
    "CouplerCirc",
    "BS",
    "Fiber",
    "PolCtrl",
    "VOA",
    "PM",
    "MZM",
    "IQM",
    "OSA",
    "ESA",
    "AWG",
    "Scope",
    "OPM",
    "PD",
    "LD",
    "Circle",
    "Square",
    # "Arrow",
    # "Line",
    "Filter",
    "Amp",
    "Circulator",
    "Adc",
    "Dac",
    "MUX",
    "FBG",
    # "ArcZ",z
    "Termination",
    "Mirror",
    "Isolator",
]

schemdraw.config(lw=1.3)
display(drawElements(elems, d="right"))
```
